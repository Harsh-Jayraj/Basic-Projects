### importing required libraries

```python
import numpy as np
import pandas as pd

import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

### defining a function that opens the link and does the html parsing using beautiful soap

```python
def getHTMLContent(link):
    html = urlopen(link)
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

### checking the contents of the site using methods of beautiful soap

```python
content = getHTMLContent('http://quotes.toscrape.com')
tables = content.find_all('div')
for table in tables:
    print(table.prettify())
```

### making a list of links of all pages of the website

```python
i=0
current_page = (content.find('li',class_='next')).a.get('href')
url='http://quotes.toscrape.com'
pages_link=['http://quotes.toscrape.com',url+current_page]
while bool(content.find('li',class_='next'))==True:
    try:
        content=getHTMLContent(url+current_page)
        current_page = (content.find('li',class_='next')).a.get('href')
        pages_link.append(url+current_page)
        print(pages_link[i])
        i+=1
    except AttributeError:
        break
print(pages_link)    
```

### dividing the contents of the website into respective group of quotes, author, and tags.

```python
all_div = content.find_all('span',class_='text')
total_div=len(all_div)
print("Total Div in the website are :",total_div)
print()
quote=[]
author=[]
tags=[]
for i in pages_link:
    cnt=content = getHTMLContent(i)
    for j in cnt.find_all('span',class_='text'):
        quote.append(j.text)
    for j in cnt.find_all('small',class_='author'):    
        author.append(j.text)
    for j in cnt.find_all('div',class_='tags'):
        tags.append(j.meta['content'])
    
print(quote)
print()
print(author)
print()
print(tags)
    
```

### now converting the divided contents into a dataframe for further use

```python
df = pd.DataFrame(list(zip(quote,author,tags)),columns = ['quote','author','tags'])
```
